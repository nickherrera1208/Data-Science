```python
import datetime 
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
```

```python
universe  = ['SPY']
```

```python
stock_data = []
for i in universe:
    try:
        stock_data.append(yf.download(i,datetime.date.today()-datetime.timedelta(3),datetime.date.today(),interval="5m", prepost=False))
    except:pass 
```

```python
stock_data[0].head()
```

```python
for i in range(len(stock_data)):
    stock_data[i]['Title'] = universe[i]
    stock_data[i]['Profit_Loss'] = float(0)
    stock_data[i]['Swing_Type'] = None
    for j in range(len(stock_data[i].High)):
        try:
            if (stock_data[i].High[j] > stock_data[i].High[j+1]) & (stock_data[i].High[j] > stock_data[i].High[j-1]):
                stock_data[i]['Swing_Type'][j] = 'Swing_High'
            elif (stock_data[i].Low[j] < stock_data[i].Low[j+1]) & (stock_data[i].Low[j] < stock_data[i].Low[j-1]):
                stock_data[i]['Swing_Type'][j] = 'Swing_Low'      
            else: stock_data[i]['Swing_Type'][j] = 'Nothing'
        except:pass
#     print(stock_data[i].isnull().sum())
stock_data[0][["High","Low","Close","Swing_Type"]]
```

```python
high_low = stock_data[0]['High'] - stock_data[0]['Low']
high_close = np.abs(stock_data[0]['High'] - stock_data[0]['Close'].shift())
low_close = np.abs(stock_data[0]['Low'] - stock_data[0]['Close'].shift())
ranges = pd.concat([high_low, high_close, low_close], axis=1)
true_range = np.max(ranges, axis=1)
atr = true_range.rolling(7).sum()/7
stock_data[0]['Half_ATR'] =atr/2
```
